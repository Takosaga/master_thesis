### Targeted Groups from Sets 

Adjusting extracted targets to if the set contain a targeted group then it counts towars the total count of each targeted group

```python
# Explode the 'extracted_target' column (convert sets into individual rows)
exploded_df = annotated_and_targeted_hatespeech_df.explode('extracted_target')

# Count the occurrences of each target group
target_counts = exploded_df['extracted_target'].value_counts()

# Plot the counts
plt.figure(figsize=(10, 6))
sns.barplot(x=target_counts.index, y=target_counts.values, palette="viridis")
plt.title("Count of Each Target Group (multiple targets in set)")
plt.xlabel("Target Group")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
print(target_counts)
print('Total of target counts for singles:',sum(target_counts))
```

### Single Targeted Groups From Sets

From sets that only target one group

```python
# Filter rows where 'extracted_target' has only one element in the set
single_target_df = annotated_and_targeted_hatespeech_df[
    annotated_and_targeted_hatespeech_df['extracted_target'].apply(lambda x: len(x) == 1)
]

single_target_df['extracted_target'] = single_target_df['extracted_target'].apply(lambda x: list(x)[0])

# Count the occurrences of each target group
target_counts = single_target_df['extracted_target'].value_counts()
```

```python
single_target_df
```

```python
print(target_counts)
print('Total of target counts for singles:',sum(target_counts))
```

```python

# Plot the counts
plt.figure(figsize=(10, 6))
sns.barplot(x=target_counts.index, y=target_counts.values, palette="viridis")
plt.title("Count of Each Target Group (single target in set)")
plt.xlabel("Target Group")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
for col in columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=col, data=single_target_df, order=single_target_df[col].value_counts().index, palette="viridis")
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Group by 'original_dataset' or 'platform' and create separate plots
grouping_columns = ['original_dataset_title', 'platform']  # Replace with the actual column names in your DataFrame

for group_col in grouping_columns:
    grouped = single_target_df.groupby(group_col)
    
    for group_name, group_data in grouped:
        plt.figure(figsize=(8, 4))
        sns.countplot(
            x='extracted_target',  # Replace with the column you want to count
            data=group_data,
            order=group_data['extracted_target'].value_counts().index,
            palette="viridis"
        )
        plt.title(f"Distribution of extracted_target for {group_col}: {group_name}")
        plt.xticks(rotation=45)
        plt.xlabel("Target Group")
        plt.ylabel("Count")
        plt.show()
```

```python
different_hatespeech_labeling = ['label_hatespeech_binary_offensive_not_included', 'label_hatespeech_binary_offensive_included'
          , 'label_normal_offensive_hatespeech']
```

```python
for label in different_hatespeech_labeling:
    for group_col in grouping_columns:
        grouped = single_target_df.groupby(group_col)
        
        for group_name, group_data in grouped:
            plt.figure(figsize=(8, 4))
            sns.countplot(
                x=label,  # Replace with the column you want to count
                data=group_data,
                order=group_data[label].value_counts().index,
                palette="viridis"
            )
            plt.title(f"Distribution of {label} for {group_col}: {group_name}")
            plt.xticks(rotation=45)
            plt.xlabel("Target Group")
            plt.ylabel("Count")
            plt.show()
```
