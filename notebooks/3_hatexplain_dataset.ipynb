```python
def process_annotations(df):
    """
    Process annotations to create:
    1. A 'label' column containing the most common label
    2. A 'target' column containing targets that appear in at least 2 annotations
    
    Args:
        df: DataFrame with an 'annotators' column containing list of annotation dictionaries
        
    Returns:
        DataFrame with two new columns: 'label' and 'target'
    """
    import pandas as pd
    from collections import Counter
    
    # Create a copy to avoid modifying the original
    result_df = df.copy()
    
    def extract_majority_label(annotators):
        # Count occurrences of each label
        labels = [a['label'] for a in annotators]
        label_counter = Counter(labels)
        
        # Find the most common label(s)
        most_common_count = label_counter.most_common(1)[0][1]
        most_common_labels = [label for label, count in label_counter.items() 
                             if count == most_common_count]
        
        # Return the first one in case of a tie
        return most_common_labels[0]
    
    def extract_common_targets(annotators):
        # Flatten the list of targets from all annotators
        all_targets = []
        for a in annotators:
            if 'target' in a and a['target']:
                all_targets.extend(a['target'])
        
        # Count occurrences of each target
        target_counter = Counter(all_targets)
        
        # Keep only targets that appear at least twice
        common_targets = [target for target, count in target_counter.items() 
                         if count >= 2]
        
        return common_targets
    
    # Apply the functions to create new columns
    result_df['label'] = result_df['annotators'].apply(extract_majority_label)
    result_df['target'] = result_df['annotators'].apply(extract_common_targets)
    print(result_df.shape)
    return result_df
```

```python
df_processed = process_annotations(df_transposed)
```
